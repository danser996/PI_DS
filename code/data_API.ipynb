# Importamos librerias requeridas

```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
from datetime import datetime
```

# Creamos un dataframe con lo requerido en los endpoints de la API

```python
file_csv = '../Dataset_def/data_movies_full.csv'
csv = pd.read_csv(file_csv, index_col=0)
```

```python
csv
```

```python
csv.columns.values
```

Creamos un nuevo dataframe solo con la informacion requerida para las respuesta de la API

```python
data = csv[['id', 'title', 'release_date', 'release_year', 'popularity', 'vote_average', 'vote_count',
                'name_actor', 'name_director', 'return']]
```

```python
# Eliminamos los duplicados de dataframe
data.drop_duplicates(inplace=True)
data.reset_index(inplace=True)
data.drop(columns='index', inplace=True)
data
```

# Creamos el campo de mes y dia en español

Ponemos el campo release_date como tipo date

```python
# convertimos la columna a tipo fecha
data.release_date = pd.to_datetime(data.release_date)
```

```python
data['release_month'] = data.release_date.apply(lambda fecha:fecha.strftime("%B"))
data.head()
```

```python
# reemplazamos los nombres de los meses en ingles a español
meses = {
    'January': 'enero',
    'February': 'febrero',
    'March': 'marzo',
    'April': 'abril',
    'May': 'mayo',
    'June': 'junio',
    'July': 'julio',
    'August': 'agosto',
    'September': 'septiembre',
    'October': 'octubre',
    'November': 'noviembre',
    'December': 'diciembre'
}

data.release_month.replace(meses, inplace=True)
data.head()
```

```python
data['release_day'] = data.release_date.apply(lambda fecha:fecha.strftime("%A"))
data.head()
```

```python
dias_semana = {
    'Monday': 'lunes',
    'Tuesday': 'martes',
    'Wednesday': 'miércoles',
    'Thursday': 'jueves',
    'Friday': 'viernes',
    'Saturday': 'sábado',
    'Sunday': 'domingo'
}

data.release_day.replace(dias_semana, inplace=True)
data.head()
```

# Reorganizamos el orden de las columnas del dataframe

```python
data = data[['id', 'title', 'release_year', 'release_month', 'release_day',
            'popularity', 'vote_average', 'vote_count', 'name_actor', 'name_director', 'return']]
data
```

```python
data.to_csv('data_API.csv')
```
