# Importamos librerias requeridas

```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
from datetime import datetime
```

# Cargamos todos los datos transformados

Cargamos el dataset transformado movies

```python
file_movies = '../Dataset_def/dtmovies.csv'
movies = pd.read_csv(file_movies, index_col=0)
movies.head(2)
```

Cargamos el dataset transformado cast

```python
file_cast = '../Dataset_def/dtcast.csv'
cast = pd.read_csv(file_cast, index_col=0)
cast.drop_duplicates(inplace=True)
cast.reset_index(inplace=True)
cast.drop(columns='index', inplace=True)
cast.head()
```

Cargamos el dataset transformado crew

```python
file_crew = '../Dataset_def/dtcrew.csv'
crew = pd.read_csv(file_crew, index_col=0)
# ya que los endpoints de la api requiere solo el director de las peliculos, descartamos los trabajos diferentes a este
crew = crew[crew.job == 'Director']
crew.drop_duplicates(inplace=True) # eliminamos los duplicados del dataframe
crew.reset_index(inplace=True) # reiniciamos el index para que queden nuevamente en orden
# ya sabemos que todos los nombres existentes corresponden a los directores
crew.drop(columns=['index', 'job'], inplace=True) # se elimina la columna job
crew.rename(columns={'name':'name_director'}, inplace=True) # reemplazamos el nombre de la columna name por name_director
crew.head()
```

# Procedemos a hacer el merge para obtener un solo dataframe

Hacemos un left join entre movies, cast y crew mediante el campo id para agregar el nombre de los actores y directores a las peliculas existentes en el dataframe de movies

```python
merge_df = pd.merge(movies, cast, on='id', how='left')
merge_df = pd.merge(merge_df, crew, on='id', how='left')
```
